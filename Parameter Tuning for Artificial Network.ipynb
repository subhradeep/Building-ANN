### Evaluating,Tunning and Improving the ANN

#### Loading the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('../Building-ANN/Churn_Modelling.csv')
```

```python
df.head()
```

### Preprocessing Steps 

```python
#Categorical variable to dummy variable
df['Geography'] = pd.get_dummies(df['Geography'],drop_first=True)
df['Gender'] = pd.get_dummies(df['Gender'],drop_first=True)

X = df.iloc[:,3:13].values
y = df.iloc[:,13].values

#Splitting into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

#Scaling the data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

### Evaluating with K-Fold Cross Validation 

Keras Classifier:

build_fn: should construct, compile and return a Keras model, which will then be used to fit/predict. Can also be passed with a function.

n_jobs=-1: Take all CPU

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense


def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
```

#### Implementing Cross Validation Method

```python
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10,n_jobs=1)
```

```python
print('Mean Accuracy of the model is: ',accuracies.mean())
print('Variance is : ',accuracies.std())
```

### Best Parameters using GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
classifier = KerasClassifier(build_fn = build_classifier)
parameters = {'batch_size':[10,25,32],
             'nb_epoch':[10,100,500]
             ,'optimizer':['adam','rmsprop']}
grid_search = GridSearchCV(estimator=classifier,
                          param_grid=parameters,
                          scoring='accuracy',
                          cv=10)
grid_search = grid_search.fit(X_train,y_train)
best_parameters = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
print('Best Parameters is :',best_parameters)
print('Best Score is:',best_score)
```
